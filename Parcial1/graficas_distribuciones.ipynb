```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu, sigma = 0, 1
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

plt.plot(x, y, label=f"Normal(μ={mu}, σ={sigma})")
plt.title("Distribución Normal")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n, p = 10, 0.4
k = np.arange(0, n+1)
y = binom.pmf(k, n, p)

plt.bar(k, y, label=f"Binomial(n={n}, p={p})")
plt.title("Distribución Binomial")
plt.xlabel("k")
plt.ylabel("P(X = k)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

N, K, n = 50, 10, 5  # Ejemplo
k = np.arange(0, n+1)
y = hypergeom.pmf(k, N, K, n)

plt.bar(k, y, label=f"Hipergeométrica(N={N}, K={K}, n={n})")
plt.title("Distribución Hipergeométrica")
plt.xlabel("k")
plt.ylabel("P(X = k)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lam = 9
k = np.arange(0, 15)
y = poisson.pmf(k, lam)

plt.bar(k, y, label=f"Poisson(λ={lam})")
plt.title("Distribución Poisson")
plt.xlabel("k")
plt.ylabel("P(X = k)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

a, b = 0, 1
x = np.linspace(a, b, 100)
y = uniform.pdf(x, loc=a, scale=b-a)

plt.plot(x, y, label=f"Uniforme([{a}, {b}])")
plt.title("Distribución Uniforme")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

alpha, beta_param = 2, 5
x = np.linspace(0, 1, 100)
y = beta.pdf(x, alpha, beta_param)

plt.plot(x, y, label=f"Beta({alpha}, {beta_param})")
plt.title("Distribución Beta")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

lmbda = 1
x = np.linspace(0, 5, 1000)
y = lmbda * np.exp(-lmbda * x)

plt.plot(x, y, label=f"Exponencial(λ={lmbda})")
plt.title("Distribución Exponencial")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

p = 0.7
k = [0, 1]
y = bernoulli.pmf(k, p)

plt.bar(k, y, width=0.4, label=f"Bernoulli(p={p})")
plt.title("Distribución Bernoulli")
plt.xlabel("k")
plt.xticks([0, 1])
plt.ylabel("P(X = k)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto

alpha = 2.5
x_m = 1
x = np.linspace(x_m, 10, 1000)
# En scipy, pareto.pdf(x, b, scale) usa b=alpha y scale=x_m
y = pareto.pdf(x, alpha, scale=x_m)

plt.plot(x, y, label=f"Pareto(x_m={x_m}, α={alpha})")
plt.title("Distribución Pareto")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
